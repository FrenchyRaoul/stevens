```python
stopwords = set(sw.words('english'))
```

```python
bad = list()


rows = list()
for doc in documents:
    words = re.sub('[^\w]+', ' ', re.sub("'", '', doc.lower())).split()
    count = Counter(words)
    rows.append(dict(count))
    
    bad.append({word for word in dict(count).keys() if word in stopwords})
```

```python
tf = pd.DataFrame(rows, index=doc_index).fillna(0).astype(int)
tf = tf[[col for col in tf.columns if col not in stopwords]]

tf
```

```python
df = tf.astype(bool).sum()
```

```python
df
```

```python
tf_3 = tf.loc[:, df[df.ge(2)].index]
tf_3
```

```python
from numpy import log, identity
from sklearn.metrics.pairwise import cosine_similarity

df_3 = tf_3.astype(bool).sum()
idf_3 = log(12 / df_3)

tf_idf_3 = tf_3 * idf_3
```
